```python
import pandas as pd
import numpy as np

# ========================
# 1. 피처 추출 함수
# ========================

def extract_features(details):
    if not isinstance(details, dict):
        return {}
    return {
        'item_form': details.get('Item Form', 'Unknown'),
        'skin_type': details.get('Skin Type', 'Unknown'),
        'material_type_free': details.get('Material Type Free', 'Unknown'),
        'product_benefits': details.get('Product Benefits', 'Unknown'),
        'sun_protection': details.get('Sun Protection', 'Unknown'),
        'active_ingredients': details.get('Active Ingredients', 'Unknown'),
    }

# ========================
# 2. Y값 생성 함수
# ========================

def calc_y2(df):
    rating_norm = (df['average_rating'] - 1) / 4
    review_weight = np.log1p(df['rating_number']) / np.log1p(df['rating_number'].max())
    y2 = (rating_norm * 0.6) + (review_weight * 0.4)
    return y2

# ========================
# 3. 피처 추출 및 Y값 생성
# ========================

features_df = df_filtered['details'].apply(extract_features).apply(pd.Series)
df_model = pd.concat([
    df_filtered[['parent_asin', 'title', 'average_rating', 'rating_number', 'category', 'sub_category']].reset_index(drop=True),
    features_df.reset_index(drop=True)
], axis=1)

df_model['y2'] = calc_y2(df_model)

print(f"총 행 수: {len(df_model)}")
print(df_model.columns.tolist())
print(df_model.head(3).to_string())
```

```python
import pandas as pd
import numpy as np

# ========================
# 키워드 사전
# ========================
keywords = {
    # 제형
    'form_cream': ['cream', 'moisturizer', 'lotion'],
    'form_serum': ['serum', 'ampoule', 'essence'],
    'form_gel': ['gel'],
    'form_toner': ['toner', 'mist', 'spray'],
    'form_oil': ['oil'],
    'form_balm': ['balm', 'stick'],
    'form_foam': ['foam', 'mousse'],
    'form_mask': ['mask', 'sheet mask', 'pack'],
    'form_cleanser': ['cleanser', 'wash', 'scrub'],

    # 피부타입
    'skin_dry': ['dry skin', 'dry'],
    'skin_oily': ['oily skin', 'oily'],
    'skin_sensitive': ['sensitive skin', 'sensitive'],
    'skin_combination': ['combination skin', 'combination'],
    'skin_normal': ['normal skin', 'normal'],
    'skin_all': ['all skin', 'all types'],

    # 성분 특성
    'paraben_free': ['paraben free', 'paraben-free'],
    'fragrance_free': ['fragrance free', 'fragrance-free', 'unscented'],
    'vegan': ['vegan'],
    'cruelty_free': ['cruelty free', 'cruelty-free'],
    'sulfate_free': ['sulfate free', 'sulfate-free'],
    'alcohol_free': ['alcohol free', 'alcohol-free'],
    'oil_free': ['oil free', 'oil-free'],
    'mineral': ['mineral'],

    # 효능
    'moisturizing': ['moisturizing', 'moisturize', 'hydrating', 'hydrate'],
    'brightening': ['brightening', 'brighten', 'whitening'],
    'anti_aging': ['anti-aging', 'anti aging', 'wrinkle', 'firming'],
    'soothing': ['soothing', 'soothe', 'calming'],
    'exfoliating': ['exfoliating', 'exfoliate'],
    'cleansing': ['cleansing', 'cleanse', 'pore'],

    # 선케어
    'spf': ['spf'],
    'zinc_oxide': ['zinc oxide'],
    'titanium_dioxide': ['titanium dioxide'],
    'broad_spectrum': ['broad spectrum', 'uva', 'uvb'],
    'water_resistant': ['water resistant', 'waterproof'],
}

# ========================
# 키워드 추출 함수
# ========================
def extract_from_features(features):
    if not features or not isinstance(features, list):
        return {k: 0 for k in keywords.keys()}
    
    text = ' '.join(features).lower()
    result = {}
    for key, kws in keywords.items():
        result[key] = 1 if any(kw in text for kw in kws) else 0
    return result

# ========================
# 테스트
# ========================
df_filtered['keyword_features'] = df_filtered['features'].apply(extract_from_features)
keyword_df = df_filtered['keyword_features'].apply(pd.Series)

print("=== 키워드별 추출 비율 ===")
for col in keyword_df.columns:
    ratio = keyword_df[col].mean() * 100
    print(f"{col}: {ratio:.1f}%")
```

```python
def fill_unknown(row):
    text = ''
    if isinstance(row['features'], list):
        text += ' '.join(row['features']).lower()
    text += ' ' + str(row['title']).lower()
    
    # item_form 채우기
    if row['item_form'] == 'Unknown':
        if 'serum' in text: row['item_form'] = 'Serum'
        elif 'gel' in text: row['item_form'] = 'Gel'
        elif 'cream' in text: row['item_form'] = 'Cream'
        elif 'lotion' in text: row['item_form'] = 'Lotion'
        elif 'toner' in text: row['item_form'] = 'Toner'
        elif 'oil' in text: row['item_form'] = 'Oil'
        elif 'foam' in text: row['item_form'] = 'Foam'
        elif 'balm' in text: row['item_form'] = 'Balm'
        elif 'mask' in text: row['item_form'] = 'Mask'
        elif 'cleanser' in text or 'wash' in text: row['item_form'] = 'Cleanser'

    # skin_type 채우기
    if row['skin_type'] == 'Unknown':
        types = []
        if 'sensitive' in text: types.append('Sensitive')
        if 'dry' in text: types.append('Dry')
        if 'oily' in text: types.append('Oily')
        if 'combination' in text: types.append('Combination')
        if 'normal' in text: types.append('Normal')
        if 'all skin' in text or 'all types' in text: types.append('All')
        if types:
            row['skin_type'] = ', '.join(types)

    # material_type_free 채우기
    if row['material_type_free'] == 'Unknown':
        free = []
        if 'paraben free' in text or 'paraben-free' in text: free.append('Paraben Free')
        if 'fragrance free' in text or 'fragrance-free' in text: free.append('Fragrance Free')
        if 'vegan' in text: free.append('Vegan')
        if 'cruelty free' in text or 'cruelty-free' in text: free.append('Cruelty Free')
        if 'sulfate free' in text or 'sulfate-free' in text: free.append('Sulfate Free')
        if 'alcohol free' in text or 'alcohol-free' in text: free.append('Alcohol Free')
        if 'oil free' in text or 'oil-free' in text: free.append('Oil Free')
        if free:
            row['material_type_free'] = ', '.join(free)

    # product_benefits 채우기
    if row['product_benefits'] == 'Unknown':
        benefits = []
        if 'moisturizing' in text or 'hydrating' in text: benefits.append('Moisturizing')
        if 'brightening' in text or 'brighten' in text: benefits.append('Brightening')
        if 'anti-aging' in text or 'anti aging' in text or 'wrinkle' in text: benefits.append('Anti-Aging')
        if 'soothing' in text or 'calming' in text: benefits.append('Soothing')
        if 'exfoliating' in text: benefits.append('Exfoliating')
        if benefits:
            row['product_benefits'] = ', '.join(benefits)

    return row

df_filtered = df_filtered.apply(fill_unknown, axis=1)

# Unknown 개수 다시 확인
for col in ['item_form', 'skin_type', 'material_type_free', 'product_benefits']:
    unknown = (df_filtered[col] == 'Unknown').sum()
    print(f"{col}: Unknown {unknown}개 ({unknown/len(df_filtered)*100:.1f}%)")
```

```python
import re

def extract_spf_value(row):
    text = ''
    if isinstance(row['features'], list):
        text += ' '.join(row['features']).lower()
    text += ' ' + str(row['title']).lower()
    
    # SPF 숫자 추출 (예: spf 50, spf50, spf30+)
    match = re.search(r'spf\s*(\d+)', text)
    if match:
        return int(match.group(1))
    return None

df_sun['spf_value'] = df_sun.apply(extract_spf_value, axis=1)

print(f"SPF 값 있는 상품: {df_sun['spf_value'].notna().sum()}개 ({df_sun['spf_value'].notna().mean()*100:.1f}%)")
print(df_sun['spf_value'].value_counts().head(10))
```

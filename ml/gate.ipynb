```python
import pandas as pd

# 규제 성분 파일 로드
df_banned = pd.read_excel(r'C:\workspace\finalproject\data\미국 화장품 규제 성분.xlsx')
df_sunscreen = pd.read_excel(r'C:\workspace\finalproject\data\미국 선크림 승인 성분.xlsx')

print(f"규제 성분 수: {len(df_banned)}")
print(df_banned['INGR_NAME'].tolist())
print()
print(f"선크림 승인 성분 수: {len(df_sunscreen)}")
print(df_sunscreen['INGR_NAME'].tolist())
```

```python
# 성분 리스트 정리 (소문자 + 공백 제거)
banned_list = [x.strip().lower() for x in df_banned['INGR_NAME'].tolist()]
sunscreen_approved_list = [x.strip().lower() for x in df_sunscreen['INGR_NAME'].tolist()]

# 게이트 함수
def ingredient_gate(product_ingredients, is_sunscreen=False):
    ingredients = product_ingredients.strip().lower()
    
    # 1단계: 규제 성분 체크
    for banned in banned_list:
        if banned in ingredients:
            return {
                "pass": False,
                "reason": f"규제 성분 포함: {banned}"
            }
    
    # 2단계: 선크림인 경우 승인 성분 체크
    if is_sunscreen:
        approved = any(approved in ingredients for approved in sunscreen_approved_list)
        if not approved:
            return {
                "pass": False,
                "reason": "선크림 승인 성분 미포함"
            }
    
    return {"pass": True, "reason": "통과"}

# 테스트
print(ingredient_gate("Water, Chloroform, Glycerin"))           # 규제 성분 포함
print(ingredient_gate("Water, Titanium dioxide", is_sunscreen=True))   # 선크림 통과
print(ingredient_gate("Water, Glycerin", is_sunscreen=True))            # 선크림 승인 성분 없음
print(ingredient_gate("Water, Glycerin"))                               # 일반 화장품 통과
```

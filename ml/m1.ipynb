```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score, KFold
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

files = {
    "cleanser": r"C:\workspace\finalproject\data\model_data_v1\model_cleanser",
    "skincare": r"C:\workspace\finalproject\data\model_data_v1\model_skincare",
    "mask":     r"C:\workspace\finalproject\data\model_data_v1\model_mask",
    "suncare":  r"C:\workspace\finalproject\data\model_data_v1\model_suncare",
}

models = {
    "Ridge":        Ridge(alpha=1.0),
    "RandomForest": RandomForestRegressor(n_estimators=200, random_state=42, n_jobs=-1),
    "XGBoost":      XGBRegressor(n_estimators=200, learning_rate=0.05, random_state=42, verbosity=0, n_jobs=-1),
    "LightGBM":     LGBMRegressor(n_estimators=200, learning_rate=0.05, random_state=42, n_jobs=-1, verbose=-1),
}

kf = KFold(n_splits=5, shuffle=True, random_state=42)

for cat, path in files.items():
    X = pd.read_csv(f"{path}_X.csv")
    y = pd.read_csv(f"{path}_y.csv").squeeze()
    
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    print(f"\n{'='*55}")
    print(f"[{cat.upper()}]")
    print(f"{'='*55}")
    print(f"{'모델':<15} {'RMSE':>10} {'MAE':>10} {'R²':>10}")
    print(f"{'-'*55}")
    
    for name, model in models.items():
        rmse = -cross_val_score(model, X_scaled, y, cv=kf, scoring='neg_root_mean_squared_error').mean()
        mae  = -cross_val_score(model, X_scaled, y, cv=kf, scoring='neg_mean_absolute_error').mean()
        r2   =  cross_val_score(model, X_scaled, y, cv=kf, scoring='r2').mean()
        print(f"{name:<15} {rmse:>10.4f} {mae:>10.4f} {r2:>10.4f}")
```
